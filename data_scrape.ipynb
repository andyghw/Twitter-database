```python
# This function takes the 'created_at', 'text', 'screen_name' and 'tweet_id' and stores it
# into a PostgreSQL database
def store_tweets_fromthepublic(tweet_id,created_at, text, screen_name,location,favorite_count,retweet_count):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWD, host=HOST, port=PORT)
    cursor = db.cursor()
    insert_query = "INSERT INTO tweets_fromthepublic(tweet_id,screen_name,created_at,location,text,favorite_count,retweet_count) VALUES ( %s, %s, %s, %s, %s, %s, %s)"
    cursor.execute(insert_query, (tweet_id,screen_name, created_at,location,text,favorite_count,retweet_count))
    db.commit()
    cursor.close()
    db.close()
    return

def store_public_accounts(screen_name,name,location):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWD, host=HOST, port=PORT)
    cursor = db.cursor()
    insert_query = "INSERT INTO public_accounts(screen_name,name,location) VALUES ( %s, %s, %s)"
    cursor.execute(insert_query, (screen_name,name,location))
    db.commit()
    cursor.close()
    db.close()
    return

def store_tags(tag,airline,tweet_id):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWD, host=HOST, port=PORT)
    cursor = db.cursor()
    insert_query = "INSERT INTO tags(tag,airline,tweet_id) VALUES ( %s, %s,%s)"
    cursor.execute(insert_query, (tag,airline,tweet_id))
    db.commit()
    cursor.close()
    db.close()
    return

for i in range (0,10):
    word=WORDS[i];
    airline=AIRLINES[i];
    for tweet in tweepy.Cursor(api.search,q=word).items():
        store_tweets_fromthepublic(tweet.id,tweet.created_at,tweet.text,tweet.user.screen_name,tweet.user.location,tweet.favorite_count,tweet.retweet_count)
        store_public_accounts(tweet.user.screen_name,tweet.user.name,tweet.user.location)
        for hashtag in tweet.entities.get('hashtags'):
            store_tags(hashtag.get('text'),airline,tweet.id) 
            print(hashtag.get('text'))
        print(tweet.created_at)
```
